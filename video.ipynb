```python
import matplotlib.pyplot as plt
import cv2
```

```python
vidcap = cv2.VideoCapture('./examples/FSD.mp4')
success, image = vidcap.read()
print(success)
plt.figure(figsize=(20,30))
plt.imshow(image)
plt.xticks([])
plt.yticks([])
plt.show()
count = 0
while success:
    success, image = vidcap.read()
    count += 1
```
